```python
import requests
import pandas as pd
import alpaca_trade_api as tradeapi
import json
from dotenv import load_dotenv
import os
```

```python
load_dotenv ()
```

```python
api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
# Set timeframe to '1D'
timeframe = "1D"
```

```python
start_date = pd.Timestamp("2020-01-01", tz="America/Chicago").isoformat()
end_date = pd.Timestamp("2020-9-14", tz="America/Chicago").isoformat()
```

```python
# Set the stock tickers
tickers = ["SPY", "LUV", "DIS", "AAPL", "SBUX", "WORK"]
```

```python
# Get 1 year's worth of historical data for all stocks
df_ticker = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
).df
```
